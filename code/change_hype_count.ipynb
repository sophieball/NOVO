```python
for hype in hype_list:
    dat[hype+"_1"] = dat[hype].map(lambda x: int(x>0))
```

```python
dat["num_unique_hype"] = dat[[h+"_1" for h in hype_list]].sum(axis=1)
```

```python
dat = dat.drop(columns=[h+"_1" for h in hype_list])
```

```python
# task 2
for i in range(100):
    include_hype = random.sample(hype_list, int(0.95*len(hype_list)))
    dat["num_hype_after_remove_5_"+str(i+1)] = dat[include_hype].sum(axis=1)
```

```python
for i in range(100):
    include_hype = random.sample(hype_list, int(0.90*len(hype_list)))
    dat["num_hype_after_remove_10_"+str(i+1)] = dat[include_hype].sum(axis=1)
```

```python
for i in range(100):
    include_hype = random.sample(hype_list, int(0.80*len(hype_list)))
    dat["num_hype_after_remove_20_"+str(i+1)] = dat[include_hype].sum(axis=1)
```
